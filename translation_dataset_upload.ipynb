```python
import csv
import json

# Open the CSV file in read mode
with open('newari_to_english_translation.csv', 'r') as csv_file:
    # Create a CSV reader object
    csv_reader = csv.reader(csv_file)
    next(csv_reader)
    # Create an empty list to store the JSON objects
    json_list = []

    # Loop through each row in the CSV file
    for row in csv_reader:
        # Create a dictionary object for the row
        translation = {
            'new': str(row[0]),
            'en': str(row[1])
        }
        # Add the dictionary object to the JSON list
        json_list.append(translation)
```

```python
print(json_list)
```

```python
import pandas as pd

# Create sample data with dictionaries
data = {
    'translation': json_list
}

# Create a DataFrame from the data
df = pd.DataFrame(data)
df = df.reset_index(drop=True)
# Display the DataFrame
print(df)
```

```python
from datasets import Dataset

new_eng_translation = Dataset.from_pandas(df)
```
